# Pull data from meteorlogy

```python
import requests
import pandas as pd

# Define the URL and headers
url = 'https://api.data.gov.my/weather/forecast'
headers = {
    'Authorization': 'Bearer YOUR_API_KEY',  # Replace with your actual API key if required
}

# Make the GET request
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()

    # Convert the JSON data to a pandas DataFrame
    # Assuming the JSON structure is a list of dictionaries or contains a key that holds the data
    df = pd.DataFrame(data)

    # Display the DataFrame
    print(df)

    # Save the DataFrame to a CSV file
    df.to_csv('weather_forecast.csv', index=False)

    print("Data saved to weather_forecast.csv")
else:
    # Print the error
    print(f"Failed to retrieve data: {response.status_code}")
    print(response.text)
```

# Data cleaning

```python
import pandas as pd

# Read file
weather=pd.read_csv('weather_forecast.csv', encoding= 'unicode_escape' )
weather.head()
```

```python
weather["location"] = weather["location"].str.replace('{','').str.replace('}', '')
weather.head()
```

```python
weather["location"] = weather["location"].str.replace("'location_id': ",'').str.replace("location_name':", '')
weather.head()
```

```python
weather["location"] = weather["location"].str.replace("'",'').str.replace("'", '')
weather.head()
#print(weather)
#weather.to_csv('locations.csv', index=False)
```

```python
#Split the 'location' column into 'location_id' and 'location_name' columns
weather[['location_id', 'location_name']] = weather['location'].str.split(',  ', expand=True)
# Drop the original 'location' column if desired
weather = weather.drop(columns=['location'])
weather.head()
```

```python
# Create the 'category' column using a lambda function with if-else conditions
weather['category'] = weather['location_id'].apply(lambda x: 'State' if x.startswith('S') else 'District' if x.startswith('D') else 'Town' if x.startswith('T') else 'Unknown')
weather.tail()
```
