## Task no.2: 
## Extract transmission information from column 'VARIANT ' into a new column 'TRANSMISSION'

```python
#Check data that contain keywords such as AT, A/T, M/T, CVT etc...
transmission= [
    vehicle['VARIANT'].str.contains('A/T','AT'),
    vehicle['VARIANT'].str.contains('M/T'),
    vehicle['VARIANT'].str.contains('Tiptronic'),
    vehicle['VARIANT'].str.contains('CVT')
]
```

```python

values = ['Automatic Transmission', 'Manual Transmission', 'Tiptronic','CVT']
```

```python
vehicle['TRANSMISSION'] = np.select(transmission, values, default='0')
```

## Task no.3: 
## Adjust the 'PRICE' column to display full prices in the currency value of RP

```python
#split curency and price (numerical value) using string function from pandas.Condition extract start after first space
vehicle[['CURRENCY', 'PRICE']] = vehicle['PRICE'].str.split(' ', n=1, expand=True)
```

```python
#Replace character ',' with blank space. 
vehicle['PRICE'] = vehicle['PRICE'].str.replace(',', '')
```

```python
vehicle
```

```python
# Create the conversion function
def converter(x):
    if 'juta' in x:
        return f"{(float(x.strip('juta'))*1000000):.2f}" #to remove unit and convert to number with two decimal places
    elif 'miliar' in x:
        return f"{(float(x.strip('miliar'))*1000000000):.2f}" #to remove unit and convert to number with two decimal places

# Print first 20 rows of head to see current values
print(vehicle.PRICE.tail(20))
```

```python
# Call the apply method
vehicle.PRICE = vehicle.PRICE.apply(converter)

# Print first 20 rows of head to see modified values
print(vehicle.PRICE.head(20))
```

## Task no.4: 
## a. Create new columns 'TAX RATE'

```python
#3% of the original book price if the original vehicle price is more than RP 1000000000
#6% of the original book price if the original vehicle price is more than RP 2000000000

def tax(vehicle):
    
    if (vehicle['PRICE'] > 1000000000 and vehicle['PRICE'] <= 2000000000 ):
        return '3%'
    elif (vehicle['PRICE'] > 2000000000):
        return '6%'
    elif (vehicle['PRICE'] <= 1000000000):
        return ''
```

```python
#
vehicle['TAX_RATE'] = vehicle.apply(tax, axis = 1)
vehicle
```

## b. Create new columns 'PRICE (INCLUDE TAX)'

```python
#Add 3% of the original book price if the original vehicle price is more than RP 1000000000
#Add 6% of the original book price if the original vehicle price is more than RP 2000000000

def price_tax(vehicle):
    
    if (vehicle['PRICE'] > 1000000000 and vehicle['PRICE'] <= 2000000000 ):
        return (vehicle['PRICE']+(vehicle['PRICE']/100*3))
    elif (vehicle['PRICE'] > 2000000000):
        return (vehicle['PRICE']+(vehicle['PRICE']/100*6))
    elif (vehicle['PRICE'] <= 1000000000):
        return vehicle['PRICE']
```

vehicle['PRICE_INCLUDE_TAX'] = vehicle.apply(price_tax, axis = 1)
